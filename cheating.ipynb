```python
import cv2 as cv
```

```python
kernel = np.ones((7,7),np.uint8)
img = cv.morphologyEx(img, cv.MORPH_CLOSE, kernel)
img = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)
# fig = plt.figure()
plt.imshow(img)
plt.savefig('flasche.png')
```
